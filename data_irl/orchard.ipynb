```python
import numpy as np
import matplotlib.pyplot as plt

from fractions import Fraction

BOUND = 50

plt.rcParams['figure.figsize'] = (20, 10)

rays = set()

for i in range(1, BOUND):
    for j in range(1, BOUND):
        rays.add(Fraction(i, j))

for ray in rays:
    x, y = ray.numerator, ray.denominator
    plt.plot([0, x], [0, y], c='blue', linewidth=0.25)

plt.show()
```
